# Stacked Bar Chart

```python
transmission_fuel_count = a.groupby(["fuelType", "transmission"]).size().unstack(fill_value=0)
transmission_fuel_count.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="coolwarm")
plt.title("Transmission Count by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.legend(title="Transmission")
plt.show()
```

# Line Chart

```python
average_price_per_year = a.groupby("year")["price"].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_price_per_year, x="year", y="price", marker="o")
plt.title("Average Price Per Year")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=a, x="fuelType", y="price", palette="pastel")
plt.title("Price Distribution by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```

# Horizontal Bar Graph

```python
# Calculate average price by fuel type
avg_price_by_fuel = a.groupby("fuelType")["price"].mean().reset_index()

# Create a horizontal bar graph
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_price_by_fuel, y="fuelType", x="price", palette="coolwarm")
plt.title("Average Price by Fuel Type")
plt.xlabel("Average Price")
plt.ylabel("Fuel Type")
plt.show()



```
